```python
import os
import glob
import random
import shutil

#diretorio = "/media/sda2/home/j/Downloads/tests/Recortes/"
diretorio = "E:/Backes/Segmentacao Linha Plantio CNN/Recortes/"

countImgPerBase = 200 #200 #300 #400 #500
random.seed(42)

for i in range(4):
    if i == 0:
        dataset = "Base_A"
    elif i == 1:
        dataset = "Base_B"
    elif i == 2:
        dataset = "Base_C"
    elif i == 3:
        dataset = "Base_D"

    # file list, img and mask
    files_path = glob.glob(diretorio + dataset + "/*.png") # + "_mask/*.png")
    files_names = [os.path.basename(x) for x in files_path]

    print("\ndataset:", dataset, "len(files_names):", len(files_names))
    #print("files_names:", files_names)

    if countImgPerBase > len(files_names):
        print("Error: countImgPerBase > len(files_names) - ", countImgPerBase, ">", len(files_names))
        raise ValueError("Error: countImgPerBase > len(files_names)")

    files_sampled = random.sample(files_names, countImgPerBase)
    #print("files_sampled:", files_sampled)
    print("len(files_sampled):", len(files_sampled))

    for j in range(2):
        if j == 0: # img
            path_source = diretorio + dataset + "/"
            output_dir = diretorio + "Base_E" + str(countImgPerBase) + "/"
        elif j == 1: # mask
            path_source = diretorio + dataset + "_mask/"
            output_dir = diretorio + "Base_E" + str(countImgPerBase) + "_mask/"

        print("\noutput_dir:", output_dir)
        if i == 0:
            try:
                os.makedirs(output_dir)
            except OSError as error:
                print(error)
            else:
                print("Successfully created the directory: %s" % output_dir)

        k = 1
        for f in files_sampled:
            f_path = path_source + f
            print("file:", k, f_path)
            shutil.copy(f_path, output_dir)
            k += 1

print("\nAll done!\n")
```
