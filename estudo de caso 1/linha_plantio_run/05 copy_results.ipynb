```python
import glob
import os
import shutil

def make_dirs(output_dir):
    try:
        os.makedirs(output_dir)
    except OSError as error:
        print(error)
    else:
        print("\nSuccessfully created the directory:", output_dir)

weights_path = "E:/Joao/jocival_linha_plantio_run/results/"
predict_sumary = weights_path + "predict_sumary/"

make_dirs(predict_sumary)

for i in range(8): #8
    if i == 0:
        dataset_weights = "Base_A"
    elif i == 1:
        dataset_weights = "Base_B"
    elif i == 2:
        dataset_weights = "Base_C"
    elif i == 3:
        dataset_weights = "Base_D"
    elif i == 4:
        dataset_weights = "Base_E200"
    elif i == 5:
        dataset_weights = "Base_E300"
    elif i == 6:
        dataset_weights = "Base_E400"
    elif i == 7:
        dataset_weights = "Base_E500"

    for j in range(8): #8
        if j == 0:
            dataset_predict = "Base_A"
        elif j == 1:
            dataset_predict = "Base_B"
        elif j == 2:
            dataset_predict = "Base_C"
        elif j == 3:
            dataset_predict = "Base_D"
        elif j == 4:
            dataset_predict = "Base_E200"
        elif j == 5:
            dataset_predict = "Base_E300"
        elif j == 6:
            dataset_predict = "Base_E400"
        elif j == 7:
            dataset_predict = "Base_E500"

        for k in range(3): #3
            if k == 0:
                network_result = "Unet"
            elif k == 1:
                network_result = "Linknet"
            elif k == 2:
                network_result = "PSPNet"

            base_to_work="weights_" + dataset_weights + "/predict_" + dataset_predict + '/'+ network_result

            full_path = weights_path + "predict/" + base_to_work
            path_to_copy = predict_sumary + base_to_work

            print(full_path + "")
            #shutil.copy(f, 'dest_folder')

            print("full_path:", full_path)
            print("path_to_copy:", path_to_copy)

            file = glob.glob(full_path + "/*.csv")
            print("file:", file)
            make_dirs(path_to_copy)

            shutil.copy(file[0], path_to_copy)

print("\n\nAll Done!\n")
```
